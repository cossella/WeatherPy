Analysis

Observed trend 1: Temperature increases as we get closer to the equator.
Observed trend 2: Humidity is about the same (20-100%) near the equator and away from it.
Observed trend 3: Wind speeds are about the same (under 10 mph) near the equator and away from it.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

import api_keys

from citipy import citipy

output_data_file = "output_data/cities.csv"

lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
lat_lngs = []
cities = []

lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name

    if city not in cities:
        cities.append(city)

len(cities)
```

```python
api_key = api_keys.api_key

url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key 
```

```python
city_name = []
country = []
latitude = []
longitude = []
temperature = []
humidity = []
cloudiness = []
wind_speed = []
date = []

record_number = 1

print(f"Beginning Data Retrieval")
print(f"-----------------------------")

for city in cities:  
    
    try: 
        selected = requests.get(f"{url}&q={city}").json() 
        city_name.append(selected["name"])
        country.append(selected["sys"]["country"])
        latitude.append(selected["coord"]["lat"])
        longitude.append(selected["coord"]["lon"])        
        temperature.append(selected["main"]["temp_max"])
        humidity.append(selected["main"]["humidity"])
        cloudiness.append(selected["clouds"]["all"])
        wind_speed.append(selected["wind"]["speed"])
        date.append(selected["dt"])
        recorded_city = selected["name"]
        print(f"Processing Record {record_number} | {recorded_city}")
        print(f"{url}&q={city}")
        
        record_number = record_number + 1
        
        time.sleep(1.01)
        
    except:
        print("Proceed to next city...")
    continue
```

```python
city_data_df = {"City": city,
                "Country": country,
                "Latitude": latitude,
                "Longitude": longitude,
                "Temperature": temperature,
                "Humidity": humidity,
                "Cloudiness": cloudiness,
                "Wind Speed": wind_speed,
                "Date": date}

city_data = pd.DataFrame(city_data_df)

city_data.count()
```

```python
city_data.head()
```

```python
plt.scatter(city_data["Latitude"], city_data["Temperature"], marker="o",s=15)

plt.title("City Latitude vs. Max Temperture (01/12/18)")
plt.xlabel("Latitude")
plt.ylabel("Max Temperture (F)")
plt.xlim(-80,100)
plt.ylim(-100,150)
plt.grid(True)

plt.show()

plt.savefig("CityLatitude_vs_Temp.png")
```
